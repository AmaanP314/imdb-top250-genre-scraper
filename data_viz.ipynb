```python
import matplotlib.pyplot as plt

Genre = ['Drama', 'Crime', 'Biography', 'Action', 'Adventure', 'Comedy', 'Animation', 'Horror', 'Mystery', 'Film-Noir']
Average_Rating = [8.288405797101458, 8.408823529411768, 8.282608695652172, 8.365957446808514, 8.228571428571426, 8.26086956521739, 8.262499999999998, 8.325, 8.4, 8.1]
positions = range(len(Average_Rating))

plt.figure(figsize=(12, 6))

bars = plt.bar(positions, Average_Rating)

plt.xticks(positions, Genre, rotation=0, ha='center') 

plt.ylim(8, 8.6)

plt.yticks([i / 10 for i in range(80, 85, 1)])  

plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Rating by Genre')

plt.tight_layout()
plt.savefig('average_rating_by_genre.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(Average_Rating, labels=Genre, autopct='%1.1f%%', startangle=140)
plt.title('Average Rating by Genre')
plt.axis('equal')  

plt.savefig('average_rating_by_genre_pie.png', dpi=300)
plt.show()
```
